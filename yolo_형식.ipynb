```python
# # JSON 파일이 있는 디렉토리 경로
# json_dir = 'data/back/SW'

# # YOLO 형식의 TXT 파일로 저장할 디렉토리 경로
# output_dir = 'data/all-data/SW'

# # 디렉토리가 존재하지 않으면 생성합니다.
# if not os.path.exists(output_dir):
#     os.makedirs(output_dir)

json_dir = "Z:/딸기 데이터/금실/01.데이터/1.Training/라벨링데이터/label/"
output_dir = "E:/strawberry/train/label변호나/"
```

```python
import os
import json

# 출력 디렉토리가 존재하지 않으면 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 레이블 매핑 (이 예에서는 임의의 예제로, 필요에 따라 조정할 수 있습니다)
labels = {'정식기': 0, '출뢰기': 1, '개화기': 2, '과실비대기': 3, '수확기': 4}

for json_filename in os.listdir(json_dir):
    if json_filename.endswith('.json'):
        try:
            with open(os.path.join(json_dir, json_filename), 'r', encoding='utf-8') as json_file:
                data = json.load(json_file)
        except json.JSONDecodeError as e:
            print(f"JSONDecodeError: 파일 '{json_filename}'을(를) 건너뜁니다. 오류: {e}")
            continue
        except Exception as e:
            print(f"오류: 파일 '{json_filename}'을(를) 처리할 수 없습니다. 오류: {e}")
            continue

        output_file_path = os.path.join(output_dir, os.path.splitext(json_filename)[0] + '.txt')

        # 이미지 크기를 숫자형으로 변환
        img_width = float(data['images']['width'])
        img_height = float(data['images']['height'])

        with open(output_file_path, 'w', encoding='utf-8') as txt_file:
            for item in data['annotations']:
                # '줄기' 관련 항목 제외
                if item['category_id'] in [category['id'] for category in data['categories'] if category['name'] == '줄기']:
                    continue
                
                class_id = item['category_id']
                bbox = item['bbox']
                x_center = (bbox[0] + bbox[2] / 2) / img_width
                y_center = (bbox[1] + bbox[3] / 2) / img_height
                width = bbox[2] / img_width
                height = bbox[3] / img_height

                txt_file.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")

        print(f"JSON 파일 '{json_filename}'이 YOLO 형식의 라벨 파일로 저장되었습니다.")

print("모든 JSON 파일을 변환하여 저장했습니다.")
```

```python
import os
import shutil

# 원본 디렉토리와 대상 디렉토리 경로
원본_디렉토리 = r'Z:/딸기 데이터/금실/01.데이터/1.Training/라벨링데이터/label/'
대상_디렉토리 = r'E:/strawberry/train/label원본/'

# 원본 디렉토리 내의 파일 목록 가져오기
파일_목록 = os.listdir(원본_디렉토리)

# 각 파일에 대해 작업 수행
for 파일 in 파일_목록:
    # 파일의 확장자 확인
    if 파일.endswith('.json'):
        # 파일의 전체 경로 생성
        원본_파일_경로 = os.path.join(원본_디렉토리, 파일)
        대상_파일_경로 = os.path.join(대상_디렉토리, 파일)

        # 파일을 대상 디렉토리로 이동
        shutil.move(원본_파일_경로, 대상_파일_경로)
        print(f'{파일}을(를) 이동했습니다.')
    else:
        print(f'{파일}은(는) jpg 파일이 아니므로 이동하지 않습니다.')
```
